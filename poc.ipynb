```python
NEO4J_URI = "neo4j://localhost:7687"
```

```python
import neo4j

neo4j_driver = neo4j.GraphDatabase.driver(
    NEO4J_URI,
)
```

```python
neo4j_driver.execute_query(query_="""
LOAD CSV WITH HEADERS FROM 'file:///content.csv' AS row
WITH row
MERGE (c:Content { content_id: row.content_id })
SET c.channel_id = row.channel_id,
    c.status = row.status,
    c.name = row.name,
    c.thumbnail_url = row.thumbnail_url,
    c.description = row.description,
    c.type = row.type,
    c.billing_type = row.billing_type,
    c.created_by = row.created_by,
    c.created_at = row.created_at,
    c.modified_by = row.modified_by,
    c.modified_at = row.modified_at,
    c.provider = row.provider,
    c.account_group_id = row.account_group_id,
    c.writer = row.writer,
    c.language_type = row.language_type,
    c.illustrator = row.illustrator,
    c.strengthening_capacity_types = row.strengthening_capacity_types,
    c.voice_actor = row.voice_actor,
    c.ar_point = row.ar_point,
    c.recommend_tag = row.recommend_tag,
    c.recommend_age = row.recommend_age,
    c.search_tag = row.search_tag,
    c.iptv_thumbnail_url = row.iptv_thumbnail_url,
    c.displayable_platform = row.displayable_platform,
    c.start_at = row.start_at,
    c.end_at = row.end_at,
    c.thumbnail_file_size = row.thumbnail_file_size,
    c.iptv_thumbnail_file_size = row.iptv_thumbnail_file_size,
    c.resource_prepared_platform = row.resource_prepared_platform,
    c.origin_server_type = row.origin_server_type,
    c.iptv_billing_type = row.iptv_billing_type,
    c.newly_received_at = row.newly_received_at,
    c.contract_types = row.contract_types,
    c.content_provider_id = row.content_provider_id,
    c.sale_price = row.sale_price,
    c.subject = row.subject;
""")
```

```python
result = neo4j_driver.execute_query(query_="""
match (n) where n:Content return n;
""")

result
```

```python
recommend_tags = [x.get("n").get("recommend_tag") for x in result.records]

distinct_recommend_tag = set()

for tags in recommend_tags:
    if not tags:
        continue
    tags_as_list = eval(tags)

    for tag in tags_as_list:
        distinct_recommend_tag.add(tag)

distinct_recommend_tag
```

```python
create_recommend_tag_node_query = """
    CREATE (n:RecommendTag {{name: '{name}'}})
"""

for tag in distinct_recommend_tag:
    query = create_recommend_tag_node_query.format(name=tag)
    result = neo4j_driver.execute_query(query_=query)
```

```python
neo4j_driver.execute_query(query_="""
CALL gds.graph.drop('tagging') YIELD graphName;
""")
```

```python
neo4j_driver.execute_query(query_="""
MATCH (source:Content)-[r:HAS_RECOMMEND_TAG]->(target:RecommendTag)
RETURN gds.graph.project(
  'tagging',
  source,
  target,
  {
    sourceNodeLabels: labels(source),
    targetNodeLabels: labels(target),
    relationshipType: 'HAS_RECOMMEND_TAG'
  },
  { undirectedRelationshipTypes: ['HAS_RECOMMEND_TAG'] }
)
""")
```

```python
neo4j_driver.execute_query(query_="""
CALL gds.fastRP.mutate(
  'tagging',
  {
    embeddingDimension: 4,
    iterationWeights: [0.8, 1, 1, 1],
    randomSeed: 42,
    mutateProperty: 'embedding'
  }
)
YIELD nodePropertiesWritten
""")
```

```python
neo4j_driver.execute_query(query_="""
CALL gds.knn.write(
  'tagging',
  {
    nodeProperties: ['embedding'],
    nodeLabels: ['Content', 'RecommendTag'],
    topK: 5,
    sampleRate: 1.0,
    deltaThreshold: 0.0,
    randomSeed: 42,
    concurrency: 1,
    writeProperty: 'score',
    writeRelationshipType: 'SIMILAR'
  }
)
YIELD similarityDistribution
RETURN similarityDistribution.mean AS meanSimilarity
""")
```

```python
import pprint

records, summary, keys = neo4j_driver.execute_query(query_="""
MATCH (c1:Content)-[r:SIMILAR]->(c2:Content)
RETURN c1.name AS content1, c2.name AS content2, r.score AS similarity
ORDER BY similarity DESCENDING
""")

pprint.pp(records)
```
